```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid")


# import and concat the three files with added column to indicate the source
df1 = pd.read_csv('sage_mnist.csv')
df1['source'] = 'MNIST'
df2 = pd.read_csv('sage_ppi.csv')
df2['source'] = 'PPI'
df3 = pd.read_csv('sage_reddit.csv')
df3['source'] = 'REDDIT-BINARY'

df = pd.concat([df1, df2, df3])

df.head()
```

```python
# drop all rows where test_score is zero
df = df[df['test_score'] != 0]
df.head()
```

```python
# make a box plot comparison between the three datasets based on the test_score


plt.figure(figsize=(10, 5))
sns.boxplot(hue='source', y='test_score', data=df, palette='Set3')  # set the color palette
plt.title('Comparison of F1 scores between the three datasets')
plt.ylabel('F1 Score')
plt.xlabel('Dataset')

# add legend names
plt.legend()
plt.show()
```
